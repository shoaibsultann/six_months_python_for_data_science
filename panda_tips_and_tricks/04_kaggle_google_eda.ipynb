```python
# Convert sizes in KB to MB
mask_kb = df['Size'].str.contains('k', na=False)
df.loc[mask_kb, 'Size'] = (df.loc[mask_kb, 'Size'].str.replace('k', '').astype(float) / 1024).round(2).astype(str) + 'M'

# Ensure sizes in MB end with 'M'
mask_mb = df['Size'].str.contains('M', na=False)
df.loc[mask_mb, 'Size'] = df.loc[mask_mb, 'Size'].str.replace('M', '').astype(float).round(2).astype(str) + 'M'
```
